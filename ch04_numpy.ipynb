#### **Boolean Indexing**

**Gi·∫£ s·ª≠** b·∫°n c√≥ m·ªôt m·∫£ng data ch·ª©a d·ªØ li·ªáu v√† m·ªôt m·∫£ng names t∆∞∆°ng ·ª©ng, trong ƒë√≥ c√≥ c√°c t√™n b·ªã l·∫∑p l·∫°i:

```python
names = np.array(["Bob", "Joe", "Will", "Bob", "Will", "Joe", "Joe"])
data = np.array([[ 4,  7],
                 [ 0,  2],
                 [-5,  6],
                 [ 0,  0],
                 [ 1,  2],
                 [-12, -4],
                 [ 3,  4]])
```

**üìå Ch·ªçn d√≤ng theo ƒëi·ªÅu ki·ªán t√™n**  

Mu·ªën l·∫•y t·∫•t c·∫£ d√≤ng c√≥ t√™n "Bob"

```python
data[names == "Bob"]
```

**‚úÖ Gi·∫£i th√≠ch:**
- names == "Bob" tr·∫£ v·ªÅ m·∫£ng Boolean.
- D√πng m·∫£ng n√†y ƒë·ªÉ ch·ªçn c√°c d√≤ng t∆∞∆°ng ·ª©ng trong data.

**üìå K·∫øt h·ª£p Boolean Indexing v·ªõi slicing**

Ch·ªçn t·∫•t c·∫£ d√≤ng "Bob" nh∆∞ng ch·ªâ l·∫•y c·ªôt th·ª© 2:

```python
data[names == "Bob", 1]
```

**üìå Lo·∫°i b·ªè d√≤ng c√≥ t√™n ‚ÄúBob‚Äù**

```python
data[names != "Bob"]
# ho·∫∑c
data[~(names == "Bob")]
```

**üìå G√°n bi·∫øn ƒëi·ªÅu ki·ªán v√† ƒë·∫£o ng∆∞·ª£c**

```python
cond = names == "Bob"
data[~cond]  # L·∫•y t·∫•t c·∫£ d√≤ng kh√¥ng ph·∫£i Bob
```

**üìå K·∫øt h·ª£p nhi·ªÅu ƒëi·ªÅu ki·ªán v·ªõi | (ho·∫∑c) v√† & (v√†)**

Ch·ªçn d√≤ng c√≥ t√™n l√† "Bob" ho·∫∑c "Will":

```python
mask = (names == "Bob") | (names == "Will")
data[mask]
```

**‚ö†Ô∏è L∆∞u √Ω:** Kh√¥ng d√πng and, or v·ªõi m·∫£ng NumPy ‚Äî h√£y d√πng &, |.

**üìå G√°n gi√° tr·ªã v·ªõi Boolean Indexing**

ƒê·∫∑t t·∫•t c·∫£ gi√° tr·ªã √¢m v·ªÅ 0

```python
data[data < 0] = 0
```

```python
data
```

G√°n to√†n b·ªô d√≤ng kh√¥ng ph·∫£i ‚ÄúJoe‚Äù th√†nh 7

```python
data[names != "Joe"] = 7
```

```python
data
```

#### **Fancy Indexing**

**Fancy indexing**(ch·ªâ m·ª•c ki·ªÉu ‚Äúƒë·∫∑c bi·ªát‚Äù) trong NumPy l√† c√°ch **truy xu·∫•t ph·∫ßn t·ª≠ b·∫±ng m·∫£ng ch·ªâ s·ªë nguy√™n**, thay v√¨ ch·ªâ d√πng l√°t c·∫Øt (slicing).

**L·∫•y nhi·ªÅu d√≤ng theo th·ª© t·ª± b·∫•t k√¨**

```python
arr = np.zeros((8, 4))
for i in range(8):
    arr[i] = i
```

```python
arr
```

```python
arr[[4, 3, 0, 6]] # l·∫•y d√≤ng 4, 3, 0, 6
```

**üìâ D√πng ch·ªâ s·ªë √¢m**

```python
arr[[-3, -5, -7]] # l·∫•y d√≤ng t·ª´ cu·ªëi l√™n: d√≤ng 5, 3, 1
```

**L·∫•y ph·∫ßn t·ª≠ theo t·ªça ƒë·ªô c·ª• th·ªÉ**

```python
arr = np.arange(32).reshape((8, 4))
arr[[1, 5, 7, 2], [0, 3, 1, 2]] 
# l·∫•y 4 ph·∫ßn t·ª≠ t·∫°i v·ªã tr√≠: (1, 0), (5, 3), (7, 1), (2, 2)
```

üìå L∆∞u √Ω: **Fancy indexing tr·∫£ v·ªÅ m·∫£ng 1 chi·ªÅu**, kh√¥ng ph·∫£i h√¨nh ch·ªØ nh·∫≠t nh∆∞ slicing.

**üìê C√°ch ch·ªçn v√πng h√¨nh ch·ªØ nh·∫≠t**

Mu·ªën ch·ªçn c·∫£ v√πng vu√¥ng (submatrix)? D√πng hai b∆∞·ªõc:

```python
arr[[1, 5, 7, 2]][:, [0, 3, 1, 2]] 
# ch·ªçn 4 d√≤ng tr∆∞·ªõc, sau ƒë√≥ ch·ªçn 4 c·ªôt t∆∞∆°ng ·ª©ng
```

**üí° Ghi nh·ªõ:**  
- Fancy indexing **lu√¥n t·∫°o m·∫£ng m·ªõi (copy)**.  
- Khi g√°n gi√° tr·ªã b·∫±ng fancy indexing, ch·ªâ nh·ªØng ph·∫ßn t·ª≠ ch·ªâ ƒë·ªãnh ƒë∆∞·ª£c thay ƒë·ªïi.

```python
arr[[1, 5, 7, 2], [0, 3, 1, 2]] = 0 
# C√°c ph·∫ßn t·ª≠ t·∫°i c√°c v·ªã tr√≠ ƒë∆∞·ª£c ch·ªçn s·∫Ω th√†nh 0.
```

##### **üîç Kh√°c bi·ªát gi·ªØa Fancy Indexing v√† Slicing**

**1. C√°ch truy c·∫≠p:**
- *Slicing*: d√πng l√°t c·∫Øt li√™n t·ª•c (:), v√≠ d·ª• arr[2:5]
- *Fancy indexing*: d√πng m·∫£ng ch·ªâ s·ªë r·ªùi r·∫°c, v√≠ d·ª• arr[[2, 5, 1]]

**2. Ki·ªÉu d·ªØ li·ªáu ch·ªâ s·ªë:**
- *Slicing*: d√πng s·ªë nguy√™n, d·∫•u : ho·∫∑c b∆∞·ªõc (start:stop:step)
- *Fancy indexing*: d√πng list ho·∫∑c m·∫£ng NumPy ch·ª©a ch·ªâ s·ªë

**3. K·∫øt qu·∫£ tr·∫£ v·ªÅ:**
- *Slicing*: l√† view (tham chi·∫øu ƒë·∫øn d·ªØ li·ªáu g·ªëc)
- *Fancy indexing*: l√† copy (sao ch√©p d·ªØ li·ªáu m·ªõi)

**4. ·∫¢nh h∆∞·ªüng khi g√°n gi√° tr·ªã:**
- *Slicing*: thay ƒë·ªïi s·∫Ω ·∫£nh h∆∞·ªüng m·∫£ng g·ªëc
- *Fancy indexing*: thay ƒë·ªïi kh√¥ng ·∫£nh h∆∞·ªüng m·∫£ng g·ªëc (tr·ª´ khi g√°n tr·ª±c ti·∫øp)

**5. Th·ª© t·ª± truy xu·∫•t:**
- *Slicing*: theo th·ª© t·ª± li·ªÅn m·∫°ch, tƒÉng d·∫ßn ho·∫∑c gi·∫£m d·∫ßn n·∫øu d√πng b∆∞·ªõc √¢m
- *Fancy indexing*: ch·ªçn theo th·ª© t·ª± b·∫•t k·ª≥, th·∫≠m ch√≠ l·∫∑p l·∫°i ph·∫ßn t·ª≠

**6. S·ªë chi·ªÅu k·∫øt qu·∫£:**
- *Slicing*: gi·ªØ nguy√™n h√¨nh d·∫°ng n·∫øu c·∫Øt nhi·ªÅu chi·ªÅu
- *Fancy indexing*: c√≥ th·ªÉ tr·∫£ v·ªÅ m·∫£ng 1 chi·ªÅu n·∫øu d√πng nhi·ªÅu m·∫£ng ch·ªâ s·ªë

**7. Hi·ªáu su·∫•t:**
- *Slicing*: nhanh h∆°n v√¨ kh√¥ng t·∫°o b·∫£n sao
- *Fancy indexing*: ch·∫≠m h∆°n v√¨ c·∫ßn t·∫°o m·∫£ng m·ªõi

##### **Swapaxes (Ho√°n ƒë·ªïi tr·ª•c)**

D√πng khi mu·ªën **ho√°n ƒë·ªïi 2 chi·ªÅu b·∫•t k·ª≥** c·ªßa m·∫£ng nhi·ªÅu chi·ªÅu (kh√¥ng ch·ªâ h√†ng-c·ªôt).

C≈©ng tr·∫£ v·ªÅ **view**, kh√¥ng copy d·ªØ li·ªáu.

#### **Expressing Conditional Logic as Array Operations**

Trong NumPy, h√†m `np.where` gi·ªëng nh∆∞ c√¢u l·ªánh r√∫t g·ªçn **x if condition else y**, nh∆∞ng ho·∫°t ƒë·ªông tr√™n **to√†n b·ªô m·∫£ng** m·ªôt c√°ch nhanh ch√≥ng.

```python
xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5])
yarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5])
cond = np.array([True, False, True, True, False])

result = np.where(cond, xarr, yarr)
result
```

### **4.7 Example: Random Walks**

**Random walk** l√† m·ªôt chu·ªói c√°c b∆∞·ªõc, m·ªói b∆∞·ªõc ƒë∆∞·ª£c quy·∫øt ƒë·ªãnh **ng·∫´u nhi√™n.**
V√≠ d·ª• ƒë∆°n gi·∫£n:
- B·∫°n b·∫Øt ƒë·∫ßu t·∫°i v·ªã tr√≠ 0.
- M·ªói l·∫ßn b·∫°n tung ƒë·ªìng xu:
	- N·∫øu **m·∫∑t s·∫•p**: b·∫°n ƒëi **+1 b∆∞·ªõc**.
	- N·∫øu **m·∫∑t ng·ª≠a**: b·∫°n ƒëi **-1 b∆∞·ªõc**.

L·∫∑p l·∫°i ƒëi·ªÅu n√†y 1000 l·∫ßn ‚Üí b·∫°n c√≥ m·ªôt ‚Äúcon ƒë∆∞·ªùng‚Äù zig-zag l√™n xu·ªëng ‚Üí g·ªçi l√† **random walk**.

üëâ G·ªçi l√† ‚Äúng·∫´u nhi√™n‚Äù v√¨ m·ªói b∆∞·ªõc b·∫°n kh√¥ng bi·∫øt tr∆∞·ªõc s·∫Ω ƒëi l√™n hay ƒëi xu·ªëng.

```python
import random
position = 0
walk = [position]
nsteps = 1000
for _ in range(nsteps):
    step = 1 if random.randint(0, 1) else -1
    position += step
    walk.append(position)
```

‚Üí M·ªói b∆∞·ªõc t·∫°o ra ng·∫´u nhi√™n s·ªë +1 ho·∫∑c -1, c·ªông d·ªìn l·∫°i ƒë·ªÉ c√≥ chu·ªói v·ªã tr√≠.

```python
plt.plot(walk[:100])  # V·∫Ω 100 b∆∞·ªõc ƒë·∫ßu ti√™n
```

**S·ª≠ d·ª•ng NumPy ƒë·ªÉ t·ªëi ∆∞u**

Ta thay th·∫ø v√≤ng l·∫∑p b·∫±ng c√°ch d√πng NumPy ƒë·ªÉ:
- Sinh ng·∫´u nhi√™n 1000 b∆∞·ªõc c√πng l√∫c (0 ho·∫∑c 1).
- D√πng np.where() ƒë·ªÉ chuy·ªÉn 0 th√†nh +1, 1 th√†nh ‚Äì1.
- T√≠nh t·ªïng c·ªông d·ªìn b·∫±ng .cumsum().

```python
nsteps = 1000
rng = np.random.default_rng(seed=12345)
draws = rng.integers(0, 2, size=nsteps)
steps = np.where(draws == 0, 1, -1)
walk = steps.cumsum()
```

‚Üí walk l√† chu·ªói v·ªã tr√≠ c·ªßa random walk.

**üìä Th·ªëng k√™:**

```python
walk.min()  
walk.max()  
```

**T√¨m th·ªùi ƒëi·ªÉm v∆∞·ª£t ng∆∞·ª°ng**

Ta mu·ªën bi·∫øt b∆∞·ªõc n√†o ƒë·∫ßu ti√™n m√† qu√° tr√¨nh v∆∞·ª£t qu√° ¬±10:

```python
(np.abs(walk) >= 10).argmax() 
```

L∆∞u √Ω: `argmax()` tr·∫£ v·ªÅ ch·ªâ s·ªë ƒë·∫ßu ti√™n c·ªßa True trong m·∫£ng boolean.

**M√¥ ph·ªèng nhi·ªÅu random walks c√πng l√∫c**

ƒê·ªÉ m√¥ ph·ªèng 5000 random walks song song (m·ªói c√°i 1000 b∆∞·ªõc), ta d√πng ma tr·∫≠n:

```python
nwalks = 5000
nsteps = 1000
draws = rng.integers(0, 2, size=(nwalks, nsteps))
steps = np.where(draws > 0, 1, -1)
walks = steps.cumsum(axis=1)  # C·ªông d·ªìn theo t·ª´ng h√†ng
```

M·ªói d√≤ng trong walks l√† m·ªôt random walk kh√°c nhau.

**üìä Th·ªëng k√™ to√†n t·∫≠p:**

```python
walks.max()  
walks.min()  
```

**Ph√¢n t√≠ch m·ª©c ƒë·ªô v∆∞·ª£t ng∆∞·ª°ng ¬±30**
